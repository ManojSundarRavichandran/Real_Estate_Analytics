Connect to SQLite3

```python
import sqlite3

conn = sqlite3.connect('realestate_data.db')
cursor = conn.cursor()

print("Connected to SQLite3 database: realestate_data.db")
```

Schema Creation

```python
listings_cleaned.columns
```

```python
create_listing_tables="""
CREATE TABLE IF NOT EXISTS listings (
    listing_id TEXT PRIMARY KEY,
    city TEXT,
    property_type TEXT,
    price REAL,
    sqft REAL,
    date_listed DATE,
    agent_id TEXT,
    latitude REAL,
    longitude REAL
);
"""
cursor.execute(create_listing_tables)
print("Listings table created successfully.")
```

```python
properties_cleaned.columns
```

```python
create_properties_table = """
CREATE TABLE IF NOT EXISTS properties (
    attribute_id INTEGER PRIMARY KEY,
    listing_id INTEGER NOT NULL,
    bedrooms INTEGER,
    bathrooms REAL,
    floor_number INTEGER,
    total_floors INTEGER,
    year_built INTEGER,
    is_rented TEXT,
    tenant_count INTEGER,
    furnishing_status TEXT,
    metro_distance_km REAL,
    parking_available TEXT,
    power_backup TEXT,
    FOREIGN KEY (listing_id) REFERENCES listings(listing_id)
)
"""

cursor.execute(create_properties_table)
print("Properties table created successfully!")
```

```python
agents_cleaned.columns
```

```python
create_agents_table = """
CREATE TABLE IF NOT EXISTS agents (
    agent_id TEXT PRIMARY KEY,
    name TEXT NOT NULL,
    phone TEXT,
    email TEXT,
    commission_rate REAL,
    deals_closed INTEGER,
    rating REAL,
    experience_years INTEGER,
    avg_closing_days REAL
)
"""

cursor.execute(create_agents_table)
print("Agents table created successfully!")
```

```python
buyers_cleaned.columns
```

```python
create_buyers_table = """
CREATE TABLE IF NOT EXISTS buyers (
    buyer_id TEXT PRIMARY KEY,
    sale_id TEXT NOT NULL,
    buyer_type TEXT,
    payment_mode TEXT,
    loan_taken TEXT,
    loan_provider TEXT,
    loan_amount REAL,
    FOREIGN KEY (sale_id) REFERENCES sales(sale_id)
)
"""

cursor.execute(create_buyers_table)
print("Buyers table created successfully!")
```

```python
sales_cleaned.columns
```

```python
create_sales_table = """
CREATE TABLE IF NOT EXISTS sales (
    sale_id INTEGER PRIMARY KEY AUTOINCREMENT,
    listing_id INTEGER NOT NULL,
    sale_price REAL,
    date_sold DATE,
    days_on_market INTEGER,
    FOREIGN KEY (listing_id) REFERENCES listings(listing_id)
)
"""

cursor.execute(create_sales_table)
print("Sales table created successfully!")
```

```python

listings_cleaned.to_sql('listings', conn, if_exists='append', index=False)
print(f"Inserted {len(listings_cleaned)} rows into listings table")

properties_cleaned.to_sql('properties', conn, if_exists='append', index=False)
print(f"Inserted {len(properties_cleaned)} rows into properties table")

agents_cleaned.to_sql('agents', conn, if_exists='append', index=False)
print(f"Inserted {len(agents_cleaned)} rows into agents table")

buyers_cleaned.to_sql('buyers', conn, if_exists='append', index=False)
print(f"Inserted {len(buyers_cleaned)} rows into buyers table")

sales_cleaned.to_sql('sales', conn, if_exists='replace', index=False)
print(f"Inserted {len(sales_cleaned)} rows into sales table")



```

Data Analysis with SQL Queries

1. What is the average listing price by city?

```python
query="""
SELECT city,AVG(price) AS avg_price
FROM listings GROUP BY city; """

avg_price_by_city = pd.read_sql_query(query, conn)
print(avg_price_by_city)
```

2. What is the average price per square foot by property type?

```python
query="""
SELECT property_type,AVG(price) AS avg_price
FROM listings GROUP BY property_type; """

avg_price_by_property_type = pd.read_sql_query(query, conn)
print(avg_price_by_property_type)
```

3.  How does furnishing status impact property prices ?

```python
query="""
SELECT 
    p.furnishing_status,
    AVG(l.price) AS avg_price
FROM listings l
JOIN properties p
    ON l.listing_id = p.listing_id
WHERE p.furnishing_status IS NOT NULL
GROUP BY p.furnishing_status;
"""
avg_price = pd.read_sql_query(query, conn)
print(avg_price)


```

4. Do properties closer to metro stations command higher prices?

```python
query = """
SELECT 
    CASE 
        WHEN p.metro_distance_km <= 1 THEN 'Near Metro (<=1 km)'
        ELSE 'Far From Metro (>1 km)'
    END AS metro_category,
    AVG(l.price) AS avg_price,
    COUNT(*) AS total_properties
FROM listings l
JOIN properties p 
    ON l.listing_id = p.listing_id
WHERE p.metro_distance_km IS NOT NULL
GROUP BY metro_category;
"""

metro_price_analysis = pd.read_sql_query(query, conn)

print(metro_price_analysis)
```

No properties closer to metro station not higher prices in the dataset 

5. Are rented properties priced differently from non-rented ones?

```python
query="""SELECT 
    p.is_rented,
    AVG(l.price) AS avg_price,
    COUNT(*) AS total_properties
FROM listings l
JOIN properties p
    ON l.listing_id = p.listing_id
WHERE p.is_rented IS NOT NULL
GROUP BY p.is_rented;
"""
rental_price_analysis = pd.read_sql_query(query, conn)
print(rental_price_analysis)
```

```python
query = """
SELECT 
    p.bedrooms,
    p.bathrooms,
    AVG(l.price) AS avg_price,
    COUNT(*) AS total_properties
FROM listings l
JOIN properties p
    ON l.listing_id = p.listing_id
WHERE p.bedrooms IS NOT NULL 
  AND p.bathrooms IS NOT NULL
GROUP BY p.bedrooms, p.bathrooms
ORDER BY p.bedrooms, p.bathrooms;
"""

bed_bath_price_analysis = pd.read_sql_query(query, conn)
print(bed_bath_price_analysis)
```

Bedrooms and bathrooms do not significantly affect property prices in this dataset

7. Do properties with parking and power backup sell at higher prices?

```python
query = """
SELECT 
    p.parking_available,
    p.power_backup,
    AVG(l.price) AS avg_price,
    COUNT(*) AS total_properties
FROM listings l
JOIN properties p
    ON l.listing_id = p.listing_id
WHERE p.parking_available IS NOT NULL
  AND p.power_backup IS NOT NULL
GROUP BY p.parking_available, p.power_backup
ORDER BY p.parking_available, p.power_backup;
"""

parking_power_analysis = pd.read_sql_query(query, conn)
print(parking_power_analysis)
```

Properties with parking and power backup have slightly higher average prices

8. How does year built influence listing price?

```python
query = """
SELECT 
    p.year_built,
    AVG(l.price) AS avg_price
FROM listings l
JOIN properties p
    ON l.listing_id = p.listing_id
WHERE p.year_built IS NOT NULL
GROUP BY p.year_built
ORDER BY p.year_built;
"""

year_price_analysis = pd.read_sql_query(query, conn)
print(year_price_analysis)

```

Listing prices do not show a consistent increase with newer construction years, indicating that year built does not significantly influence pricing in this dataset.

```python
query = """
SELECT 
    CASE 
        WHEN price < 2000000 THEN 'Low Price'
        WHEN price < 2500000 THEN 'Medium Price'
        ELSE 'High Price'
    END AS price_group,
    COUNT(*) AS total_properties
FROM listings
GROUP BY price_group;
"""

price_distribution = pd.read_sql_query(query, conn)
print(price_distribution)
```

What is the average days on market by city?

```python
query = """
SELECT 
    l.city,
    AVG(s.days_on_market) AS avg_days_on_market
FROM listings l
JOIN sales s
    ON l.listing_id = s.listing_id
WHERE s.days_on_market IS NOT NULL
GROUP BY l.city
ORDER BY avg_days_on_market DESC;
"""

avg_days_market = pd.read_sql_query(query, conn)
print(avg_days_market)
```

Los Angeles has the highest average days on market,Houston has the lowest average

Which property types sell the fastest?

```python
query = """
SELECT 
    l.property_type,
    AVG(s.days_on_market) AS avg_days_on_market
FROM listings l
JOIN sales s
    ON l.listing_id = s.listing_id
WHERE s.days_on_market IS NOT NULL
GROUP BY l.property_type
ORDER BY avg_days_on_market ASC;
"""

fastest_selling_types = pd.read_sql_query(query, conn)
print(fastest_selling_types)
```

Houses sell faster, condo takes longer time to sell.

```python

query = """
SELECT 
    ROUND(
        100.0 * SUM(CASE WHEN s.sale_price > l.price THEN 1 ELSE 0 END) 
        / COUNT(*), 2
    ) AS percentage_above_listing
FROM listings l
JOIN sales s
    ON l.listing_id = s.listing_id;
"""

percentage_above = pd.read_sql_query(query, conn)
print(percentage_above)
```

approximately 49.31% of properties were sold above their listing price.

What is the sale-to-list price ratio by city?

```python
query = """
SELECT 
    l.city,
    ROUND(AVG(s.sale_price * 1.0 / l.price), 3) AS sale_to_list_ratio
FROM listings l
JOIN sales s
    ON l.listing_id = s.listing_id
WHERE l.price IS NOT NULL 
  AND s.sale_price IS NOT NULL
GROUP BY l.city
ORDER BY sale_to_list_ratio DESC;
"""

ratio_by_city = pd.read_sql_query(query, conn)
print(ratio_by_city)
```

How does metro distance affect time on market?

What is the monthly sales trend?

```python
query = """
SELECT 
    substr(date_sold, 1, 7) AS month,
    COUNT(*) AS total_sales
FROM sales
GROUP BY month
ORDER BY month;
"""

monthly_sales = pd.read_sql_query(query, conn)
print(monthly_sales)
```

Sales grew through 2023, reached a peak in October 2023, and then gradually declined during 2024.

Which agents have closed the most sales?

```python
query = """
SELECT 
    a.agent_id,
    a.name,
    COUNT(s.listing_id) AS total_sales_closed
FROM listings l
JOIN sales s
    ON l.listing_id = s.listing_id
JOIN agents a
    ON l.agent_id = a.agent_id
GROUP BY a.agent_id, a.name
ORDER BY total_sales_closed DESC;
"""

top_agents = pd.read_sql_query(query, conn)
print(top_agents)
```

Do agents with higher ratings close deals faster?

What is the average commission earned by each agent?

```python
query = """
SELECT 
    a.agent_id,
    AVG(s.sale_price * (a.commission_rate / 100.0)) AS avg_commission
FROM sales s
JOIN listings l 
    ON s.listing_id = l.listing_id
JOIN agents a 
    ON l.agent_id = a.agent_id
GROUP BY a.agent_id
ORDER BY avg_commission DESC;
"""

result = pd.read_sql_query(query, conn)
print(result)
```

What percentage of buyers are investors vs end users?

```python
query = """
SELECT 
    buyer_type,
    COUNT(*) * 100.0 / (SELECT COUNT(*) FROM buyers) AS percentage
FROM buyers
GROUP BY buyer_type;
"""

result = pd.read_sql_query(query, conn)
print(result)
```

```python
query = """
SELECT 
    l.city,
    SUM(b.loan_taken) * 100.0 / COUNT(*) AS loan_uptake_rate
FROM buyers b
JOIN listings l 
    ON b.sale_id = l.listing_id
GROUP BY l.city
ORDER BY loan_uptake_rate DESC;
"""

result = pd.read_sql_query(query, conn)
print(result)
```

Los Angeles has the highest loan uptake, but the difference is very small.
